```python
arr = [8, 5, 2, 6, 4, 9, 1, 3, 7, 0, 10, 123, 100, 45, 67]
length = len(arr)
print("Length of the array:", length)

for i in range(length):
    print("Element at index", i, "is", arr[i])
```

```python
i = 0
for i in range(length - 1):
    for j in range(length - i - 1):
        print("Comparing elements at index", j, "and", j + 1)
        if arr[j] > arr[j + 1]:
            print("Swapping elements", arr[j], "and", arr[j + 1])
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
            print("Array after swap:", arr)
    print("Array after pass", i + 1, ":", arr)

print("Sorted array:", arr)
```
